## Further Analysis

#### What was the best month for sales? How much was earned that month?

```python
#lets add a new column to calculate the total price

data["total_price"] = data["Qty_Ordered"] * data["Price_Each"]
```

```python
data.head()
```

```python
#Add month column and extracting the month from the Order date


data["Month"] = data["Order_Date"].str[0:2]
data["Month"] = data["Month"].astype(int)
```

```python
data.head()
```

```python
#now lets answer the question

sales_by_month = data.groupby("Month")["total_price"].sum().reset_index()
sales_by_month
```

```python
sales_by_month.sort_values("total_price",ascending=False)
```

```python
# plot

month = range(1,13)

plt.bar(month,sales_by_month["total_price"], color="green")
plt.xticks(month)
plt.xlabel("Month")
plt.ylabel("Sales in Millions - USD")
plt.show()
```

#### So month of December was the best month among Sales followed by Oct and April. Total sales for month of Dec was 4608295. Least spend is on month of Jan

### What city had the highest number of sales?

```python
data.head()
```

```python
#using apply function to split the address and get only the city.

def city(value):
    value = str(value).split(',')
    value = value[1]
    return str(value)

data["City"] = data["Address"].apply(city)
data.drop("city",axis=1,inplace=True)
```

```python
data.head()
```

```python
city_sales_highest = data.groupby("City")["total_price"].sum().reset_index()
city_sales_highest.sort_values("total_price",ascending=False)
```

```python
#plot

plt.figure(figsize=(11, 5))
plt.bar(city_sales_highest["City"],city_sales_highest["total_price"], width=0.3)
plt.xlabel("City")
plt.ylabel("Sales in Millions USD")
plt.show()
```

#### So San Francisco is best  city among Sales followed by Log Angels and Ney York. City with low sales was Austin

### What time should we display advertisements to maximize the likelihood of customers buying the product?

```python
#convert the order_date to a datetime datatype

data["Order_Date"] = pd.to_datetime(data["Order_Date"])
```

```python
data.head()
```

```python
data["hour"] = data["Order_Date"].dt.hour
data["minute"] = data["Order_Date"].dt.minute
```

```python
data.head()
```

```python
orders_by_hours = data.groupby("hour")["Order_id"].count().reset_index()
orders_by_hours.sort_values("hour",ascending=True)
```

```python
#plot

plt.figure(figsize=(9,5))
plt.plot(orders_by_hours["hour"],orders_by_hours["Order_id"])
plt.xticks(orders_by_hours["hour"])
plt.xlabel("Time in 24 hours format")
plt.ylabel("No of Orders Placed")
plt.grid() #to show lines
plt.show()
```

#### More orders are placed between morning 10am till 9pm in the evening. Post 9pm orders gets dropped. Reommendation to put  ads between 11 am in the morning till 7pm in the evening 

### What products are most often sold together?

```python
data.head()
```

```python
together = data.loc[data["Order_id"].duplicated(keep=False)] #order id duplicated values are captured together. False to keep all
together
```

```python
together["Grouped"] = together.groupby("Order_id")["Product"].transform(lambda x: ','.join(x))
```

```python
together.head()
```

```python
together = together[["Order_id","Grouped"]].drop_duplicates() #delete all rows with order id and grouped the same
```

```python
together.head()
```

```python
df = together.groupby("Grouped")["Order_id"].count().reset_index()
df.sort_values("Order_id",ascending=False).head(10)
```
